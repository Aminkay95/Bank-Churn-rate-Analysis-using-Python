### 1. How do customer demographics (Gender, Age, Geography) correlate with the number of products (NumOfProducts) a customer uses?

```python
# Looking at how gender affects Number of products a customer uses

gender_grouped = df.groupby('Gender')['NumOfProducts'].mean()

gender_grouped

x = gender_grouped.index
y = gender_grouped.values


plt.bar(x,y)
plt.title('Number of Products on average bought by each gender')
plt.xlabel('Gender')
plt.ylabel('Average Number of Products used')
plt.show()
```

Based on the visuals, it appears that, on average, female customers use slightly more product than male customers.

```python
# Geography demographic

geography_grouped = df.groupby('Geography')['NumOfProducts'].mean()

geography_grouped

x = geography_grouped.index
y = geography_grouped.values


plt.bar(x,y)
plt.title('Number of Products bought by Country')
plt.xlabel('Country')
plt.ylabel('Average Number of Products used')
plt.show()
```

A customer from Spain on average uses silightly more products than from the other coutries while a customer from germany uses the least number of products on average


```python
# Age demographic

# Grouping the Data in age into age range for easier analysis

bins = [18, 35, 50, 100]
labels = ['18-35', '36-50','50+']

df['Age range'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

df
```

```python
# plotting the age demographic 

age_grouped = df.groupby('Age range')['NumOfProducts'].mean()

x = age_grouped.index
y = age_grouped.values

plt.bar(x,y)
plt.title('Number of Products on average bought by Age group')
plt.xlabel('Age group')
plt.ylabel('Average Number of Products used')
plt.show()
```

Customers in the 50+ age group use the least number of products on average while customers in the 18-35 age group use the most number of produts followed closely by those in the 36-50 group.

Customer demographics show the following trends in relation to the number of products used:

- **Gender**: On average, female customers use slightly more products than male customers.
- **Age**: Customers in the 50+ age group use the fewest products on average, while those in the 18-35 age group use the most, followed closely by customers in the 36-50 age group.
- **Geography**: Customers from Spain, on average, use slightly more products than those from other countries, while customers from Germany use the fewest.

These trends suggest that gender, age, and geography influence the number of products a customer uses, with younger customers, females, and those from Spain generally using more products.

### 2.Is there a relationship between account balance (Balance) and credit score (CreditScore)?

```python
# We use scatter plot to see if there is any relation

plt.scatter(df['CreditScore'],df['Balance'])
plt.xlabel('Credit Score')
plt.ylabel('Balance')
plt.title('Scatter plot of credit score and balance')
plt.show()
```

```python
correlation = df['CreditScore'].corr(df['Balance'])
print(correlation)
```

Based on the scatter plot and correlation coefficient, there appears to be no significant relationship between a customer's account balance and their credit score.

### 3.Does tenure (Tenure) with the bank affect customer activity (IsActiveMember)?

```python
# reassigning the data in IsActiveMember to either 1 or 0 with 1 representing an active memeber
def convert_isactive(activity):
    if activity == 'Yes':
        activity = 1
    else:
        activity = 0
    return activity
df['IsActiveMember'] = df['IsActiveMember'].apply(convert_isactive)

df
```

```python
# Using Correlation Coefficient to see if there is a relationship between the customers activity and their tenure

correlation = df['Tenure'].corr(df['IsActiveMember'])
print(correlation)
```

The tenure (Tenure) with the bank does not appear to affect customer activity (IsActiveMember). This is supported by the correlation coefficient, which is close to zero, indicating no significant relationship between tenure and activity. Since the correlation is neither close to 1 nor -1, we can conclude that there is no strong relationship between the two variables.

### 4.What is the impact of salary (EstimatedSalary) and credit card ownership (HasCrCard) on customer retention?

```python
# Looking at the impact of Salary on customer retention

bins=[0, 50000, 100000, 150000, 200000]
labels = ['0-50K','50k-100k','100k-150k','150k+']

df['Salary band'] = pd.cut(df['EstimatedSalary'], bins=bins, labels=labels, right=True)
```

```python
salary_grouped = df.groupby('Salary band')['Exited'].sum()

x = salary_grouped.index
y = salary_grouped.values

plt.bar(x,y)
plt.title('Salary band vs Number of customer churned')
plt.xlabel('Salary Band')
plt.ylabel('Number of Customer Churned')
plt.show()
```

Salary range has a weak impact on the number of churned. As we can see in the visual in the higher salary band customers tend to churn in higher numbers so we can assume that the higher the salary band the higher the chances of the customer churning

```python
# Impact of HasCrCard

cr_grouped = df.groupby('HasCrCard')['Exited'].sum()

x = cr_grouped.index
y = cr_grouped.values

plt.bar(x,y)
plt.title('Credit Card Holders vs Number of customer churned')
plt.xlabel('Is Customer Credit Card Holder? ')
plt.ylabel('Number of Customer Churned')
plt.show()
```

Of the Customers Churned we can see that most of them were not credit card holders. So we can assume that credit card ownership reduces the chances of customer getting churned. In conlusion credit card ownership has an impact on customer retention.

Salary has a weak impact on customer retention, with higher salary bands showing a tendency for higher churn rates. This suggests that customers in higher salary brackets are more likely to churn. Regarding credit card ownership, most churned customers were not credit card holders, indicating that having a credit card reduces the likelihood of churn. In conclusion, credit card ownership has a significant impact on customer retention.
